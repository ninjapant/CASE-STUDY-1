### Spark Funds wants to invest in countries with the highest amount of funding for the chosen investment type. This is a part of its broader strategy to invest where most investments are occurring.

```python
master_frame.groupby('country_code')['raised_amount_usd'].mean().sort_values(ascending=False)
```

```python
master_frame.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False).head(15)
```

```python
def verify(values):
    value=values.split('|')
    if len(value)>0:
        return value[0]
    else:
        return values
master_frame.head(10)
```

```python
master_frame['sector'] = master_frame['category_list'].apply(verify)
master_frame['sector'] = master_frame['sector'].str.lower()
```

```python
master_frame.head(10)
```

```python
top9=master_frame[master_frame["country_code"].isin(['USA', 'CHN', 'GBR', 'IND', 'CAN', 'FRA', 'ISR', 'DEU', 'JPN' ])]
```

```python
top9.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False).head()
```

```python
plt.figure(figsize=(25, 10))
plt.subplot(1, 2, 1)
sns.countplot(x="country_code", data=top9)
plt.title("Count")
plt.subplot(1, 2, 2)
sns.barplot(x="country_code", y="raised_amount_usd", data=top9, estimator=sum)
plt.title("Sum")
plt.show()
```
